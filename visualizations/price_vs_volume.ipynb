```python
import pandas as pd 
import altair as alt
import os
```

```python
os.getcwd()
```

```python
df = pd.read_csv("../data/processed/coingecko_processed.csv", parse_dates=["date"])
```

```python
btc = df[df["coin"] == "bitcoin"]
```

```python
# --- Top panel: Price with moving averages ---
price_line = alt.Chart(btc).mark_line(color="#f7931a").encode(
    x=alt.X("date:T", title="Date"),
    y=alt.Y("price:Q", title="Price (USD)", scale=alt.Scale(zero=False)),
    tooltip=["date:T", "price:Q", "daily_return:Q"]
)
```

```python
ma7_line = alt.Chart(btc).mark_line(color="white", strokeDash=[4, 2], opacity=0.6).encode(
    x="date:T",
    y="price_7d_avg:Q"
)
```

```python
ma30_line = alt.Chart(btc).mark_line(color="steelblue", strokeDash=[6, 3], opacity=0.8).encode(
    x="date:T",
    y="price_30d_avg:Q"
)
```

```python
# Highlight large swing days
swings = alt.Chart(btc[btc["large_swing"]]).mark_rule(color="red", opacity=0.4).encode(
    x="date:T",
    tooltip=["date:T", "daily_return:Q"]
)
```

```python
price_chart = (price_line + ma7_line + ma30_line + swings).properties(
    width=800,
    height=300,
    title="Bitcoin Price with Moving Averages (Red = Large Swing Days)"
)
```

```python
# --- Bottom panel: Volume bars ---
volume_bars = alt.Chart(btc).mark_bar().encode(
    x=alt.X("date:T", title="Date"),
    y=alt.Y("volume:Q", title="Volume (USD)"),
    color=alt.condition(
        alt.datum.high_volume,
        alt.value("orange"),
        alt.value("steelblue")
    ),
    tooltip=["date:T", "volume:Q", "high_volume:N"]
).properties(
    width=800,
    height=150,
    title="Daily Volume (Orange = High Volume Days)"
)
```

```python
# --- Combine vertically ---
chart = alt.vconcat(price_chart, volume_bars).configure(
    background="#1e1e1e"
).configure_axis(
    labelColor="white",
    titleColor="white",
    gridColor="#333333"
).configure_title(
    color="white"
)
```
